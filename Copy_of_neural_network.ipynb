```python
# this is more complex then previous one. It is variation of the previous code.
# I use a loop to get the weights, inputs and biases then feed into the neurons.
inputs = [1, 2, 3, 2.5]
weights = [
    [0.2, 0.8, -0.5, 1],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]
biases = [2, 3, 0.5]

# Output of the current layer
layer_outputs = []


# For each neuron
for neuron_weights, neuron_bias in zip(weights, biases):
    # Zeroed output of the given neuron
    neuron_output = 0

    # For each input and weight to the neuron
    for n_input, weight in zip(inputs, neuron_weights):
        # Multiply this input by the associated weight
        # and add to the neuron's output variable

        neuron_output += n_input * weight

    # Add bias
    neuron_output += neuron_bias

    # Put the neuron's result into the layer's output list
    layer_outputs.append(neuron_output)

#print(layer_outputs)
w_out=[0.2, 0.8, -0.5]
biase_out=0.5
final_output=(layer_outputs[0]*w_out[0]+layer_outputs[1]*w_out[1]+layer_outputs[2]*w_out[2])+biase_out
print(final_output)
```

#Example-8: Solve AND gate problem using neural network.

```python
import numpy as np

# Define the inputs and corresponding labels for the AND gate
inputs = np.array([[0, 0],
                   [0, 1],
                   [1, 0],
                   [1, 1]])

labels = np.array([0, 0, 0, 1])

# Define the weights and biases for the neural network
weights = np.array([0.5, 0.5])
bias = -0.7

# Define the activation function (step function)
def step_function(x):
    return 0 if x <= 0 else 1

# Perform the forward pass to compute the outputs
outputs = []
for input_row in inputs:
    weighted_sum = np.dot(input_row, weights) + bias
    output = step_function(weighted_sum)
    outputs.append(output)

# Compare the predicted outputs with the actual labels
correct_predictions = np.sum(outputs == labels)

# Print the results
print("Inputs:", inputs)
print("Labels:", labels)
print("Predicted Outputs:", outputs)
print("Correct Predictions:", correct_predictions)
```

## Solve OR gate problem using neural network.

```python
import numpy as np

# Define the inputs and corresponding labels for the OR gate
inputs = np.array([[0, 0],
                   [0, 1],
                   [1, 0],
                   [1, 1]])

labels = np.array([0, 1, 1, 1])  # OR gate outputs

# Define the weights and bias for the OR gate
weights = np.array([0.5, 0.5])
bias = -0.2  # Chosen to correctly model the OR gate

# Define the activation function (step function)
def step_function(x):
    return 0 if x <= 0 else 1

# Perform the forward pass to compute the outputs
outputs = []
for input_row in inputs:
    weighted_sum = np.dot(input_row, weights) + bias
    output = step_function(weighted_sum)
    outputs.append(output)

# Compare the predicted outputs with the actual labels
correct_predictions = np.sum(outputs == labels)

# Print the results
print("OR Gate:")
print("Inputs:", inputs)
print("Labels:", labels)
print("Predicted Outputs:", outputs)
print("Correct Predictions:", correct_predictions)
```

## Solve NOT gate problem using neural network.

```python
import numpy as np

# Define the inputs and corresponding labels for the NOT gate
inputs = np.array([[0],  # Single input 0
                   [1]]) # Single input 1

labels = np.array([1, 0])  # NOT gate outputs

# Define the weight and bias for the NOT gate
weight = -1  # A negative weight to invert the input
bias = 0.5   # Bias chosen to correctly model the NOT gate

# Define the activation function (step function)
def step_function(x):
    return 0 if x <= 0 else 1

# Perform the forward pass to compute the outputs
outputs = []
for input_value in inputs:
    weighted_sum = np.dot(input_value, weight) + bias
    output = step_function(weighted_sum)
    outputs.append(output)

# Compare the predicted outputs with the actual labels
correct_predictions = np.sum(outputs == labels)

# Print the results
print("\nNOT Gate:")
print("Inputs:", inputs.flatten())
print("Labels:", labels)
print("Predicted Outputs:", outputs)
print("Correct Predictions:", correct_predictions)
```

# Example-10: Calculate loss function and then minimize the error after Nth iterations by adjusting the parameters.

```python
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Use sigmoidal activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the neural network architecture
input_size = 2
hidden_size = 3
output_size = 1

# Initialize the weights and biases with a random seed
W1 = np.random.randn(input_size, hidden_size)
b1 = np.random.randn(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.random.randn(output_size)

# Define the input for the AND gate
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Define the target output for the AND gate
target = np.array([[0], [0], [0], [1]])

# Parameters for training
learning_rate = 0.1
epochs = 10000
N = 1000  # Number of iterations after which to print loss and adjust parameters
threshold = 0.00001  # Set the threshold in percentage (e.g., 0.09% or 5%)

for epoch in range(epochs):
    # Forward pass
    hidden_layer_output = sigmoid(np.dot(x, W1) + b1)
    output = sigmoid(np.dot(hidden_layer_output, W2) + b2)

    # Calculate the loss (using Mean Squared Error)
    loss = np.mean((output - target) ** 2)

    # Backpropagation
    d_output = (output - target) * output * (1 - output)  # Derivative of MSE and sigmoid
    d_hidden_layer = d_output.dot(W2.T) * hidden_layer_output * (1 - hidden_layer_output)

    # Update weights and biases
    W2 -= learning_rate * hidden_layer_output.T.dot(d_output)
    b2 -= learning_rate * np.sum(d_output, axis=0)
    W1 -= learning_rate * x.T.dot(d_hidden_layer)
    b1 -= learning_rate * np.sum(d_hidden_layer, axis=0)

    # Print loss and adjust parameters every N iterations
    if (epoch + 1) % N == 0:  # Using (epoch + 1) so it starts from the first N
        loss_percentage = loss * 100  # Convert loss to percentage
        print(f"Epoch {epoch + 1}, Loss: {loss_percentage:.2f}%")

        # Check if the loss percentage is below the threshold
        if loss_percentage < threshold:
            print(f"Early stopping as loss ({loss_percentage:.2f}%) is below the threshold ({threshold}%).")
            break

# Final output after training
print("\nFinal Output after Training:")
print(output)
loss_percentage = loss * 100  # Convert final loss to percentage
print(f"Final Loss: {loss_percentage:.2f}%")
```

# Example-11: Solving AND gate problem using neural network with backpropagation approach.

```python
import numpy as np

# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid activation function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Define the input data and labels for the AND gate problem
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [0], [0], [1]])

# Initialize the weights and biases with random values
np.random.seed(42)
W = np.random.randn(2, 1)
b = np.random.randn(1)

# Set the learning rate and number of epochs
learning_rate = 0.8
num_epochs = 500

# Training loop
for epoch in range(num_epochs):
    # Forward pass
    z = np.dot(X, W) + b
    a = sigmoid(z)

    # Compute the loss
    loss = np.mean((y - a) ** 2)

    error= (y - a)
    # Backpropagation
    delta = error * sigmoid_derivative(z)
    dW = np.dot(X.T, delta)
    db = np.sum(delta)

    # Update the weights and biases
    W -= learning_rate * dW
    b -= learning_rate * db

    # Print the loss at every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch + 1}: Loss = {loss}")

# Test the model on the input data
#predictions = (a >= 0.5).astype(int)
print("Predicted Output:")
print(a)
```

# HW: Solve OR, NOR, NAND using backpropagation algorithm

```python
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Neural network parameters
input_size = 2
hidden_size = 2
output_size = 1
learning_rate = 0.1
epochs = 10000

# Initialize weights and biases
W1 = np.random.randn(input_size, hidden_size)
b1 = np.random.randn(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.random.randn(output_size)

# Define input data for logic gates
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Define target outputs for different gates
gates = {
    "AND": np.array([[0], [0], [0], [1]]),
    "OR": np.array([[0], [1], [1], [1]]),
    "NOR": np.array([[1], [0], [0], [0]]),
    "NAND": np.array([[1], [1], [1], [0]]),
    "XOR": np.array([[0], [1], [1], [0]])
}

# Training function
def train_gate(target_output):
    global W1, b1, W2, b2  # Use global variables to update weights and biases
    for epoch in range(epochs):
        # Forward pass
        hidden_layer_input = np.dot(x, W1) + b1
        hidden_layer_output = sigmoid(hidden_layer_input)
        final_input = np.dot(hidden_layer_output, W2) + b2
        output = sigmoid(final_input)

        # Calculate the loss (Mean Squared Error)
        loss = np.mean((output - target_output) ** 2)

        # Backpropagation
        error = output - target_output
        d_output = error * sigmoid_derivative(output)
        hidden_error = d_output.dot(W2.T)
        d_hidden_layer = hidden_error * sigmoid_derivative(hidden_layer_output)

        # Update weights and biases
        W2 -= learning_rate * hidden_layer_output.T.dot(d_output)
        b2 -= learning_rate * np.sum(d_output, axis=0)
        W1 -= learning_rate * x.T.dot(d_hidden_layer)
        b1 -= learning_rate * np.sum(d_hidden_layer, axis=0)

        # Print the loss every 1000 epochs
        if (epoch + 1) % 1000 == 0:
            print(f"Epoch {epoch + 1}, Loss: {loss:.4f}")

# Function to test the trained model
def test_gate():
    hidden_layer_input = np.dot(x, W1) + b1
    hidden_layer_output = sigmoid(hidden_layer_input)
    final_input = np.dot(hidden_layer_output, W2) + b2
    output = sigmoid(final_input)
    return output

# Train and test for each logic gate
for gate, target_output in gates.items():
    print(f"\nTraining for {gate} gate:")
    train_gate(target_output)
    print(f"Results for {gate} gate:")
    print(test_gate())
```

```python
import numpy as np

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def deriv_sigmoid(x):
    fx = sigmoid(x)
    return fx * (1 - fx)

# Mean Squared Error Loss
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

class SimpleNeuralNetwork:
    def __init__(self):
        # Initialize weights and biases
        self.weights = np.random.normal(size=(2, 2))  # 2x2 weights for the hidden layer
        self.biases = np.random.normal(size=2)        # 2 biases for the hidden layer
        self.output_weights = np.random.normal(size=2)  # Weights for the output layer
        self.output_bias = np.random.normal()          # Bias for the output layer

    def feedforward(self, x):
        # Compute hidden layer outputs
        hidden_outputs = sigmoid(np.dot(x, self.weights) + self.biases)
        # Compute output layer output
        output = sigmoid(np.dot(hidden_outputs, self.output_weights) + self.output_bias)
        return output

    def train(self, data, y_true):
        learning_rate = 0.1
        epochs = 1000

        for epoch in range(epochs):
            total_loss = 0

            for x, y in zip(data, y_true):
                # Forward pass
                hidden_sum = np.dot(x, self.weights) + self.biases
                hidden_outputs = sigmoid(hidden_sum)

                output_sum = np.dot(hidden_outputs, self.output_weights) + self.output_bias
                y_pred = sigmoid(output_sum)

                # Compute gradients
                d_L_d_ypred = -2 * (y - y_pred)

                # Gradients for the output layer
                d_ypred_d_output_weights = hidden_outputs * deriv_sigmoid(output_sum)
                d_ypred_d_output_bias = deriv_sigmoid(output_sum)

                # Gradients for the hidden layer
                d_ypred_d_hidden = self.output_weights * deriv_sigmoid(output_sum)
                d_hidden_d_weights = np.outer(x, deriv_sigmoid(hidden_sum))

                # Update weights and biases
                self.output_weights -= learning_rate * d_L_d_ypred * d_ypred_d_output_weights
                self.output_bias -= learning_rate * d_L_d_ypred * d_ypred_d_output_bias
                self.weights -= learning_rate * d_L_d_ypred * d_ypred_d_hidden * d_hidden_d_weights
                self.biases -= learning_rate * d_L_d_ypred * d_ypred_d_hidden * deriv_sigmoid(hidden_sum)

            # Calculate loss for the current epoch
            y_preds = np.apply_along_axis(self.feedforward, 1, data)
            loss = mse_loss(y_true, y_preds)
            total_loss += loss

            # Print loss every 100 epochs
            if epoch % 100 == 0:
                print(f"Epoch {epoch}, Loss: {total_loss}")

# Example dataset
data = np.array([[-2, -1], [25, 6], [17, 4], [-15, -6]])
all_y_trues = np.array([1, 0, 0, 1])

# Train the neural network
network = SimpleNeuralNetwork()
network.train(data, all_y_trues)
```

```python
# source doce: https://towardsdatascience.com/machine-learning-for-beginners-an-introduction-to-neural-networks-d49f22d238f9

import numpy as np
rafi=[]
shafi=[]
def sigmoid(x):
  # Sigmoid activation function: f(x) = 1 / (1 + e^(-x))
  return 1 / (1 + np.exp(-x))

def deriv_sigmoid(x):
  # Derivative of sigmoid: f'(x) = f(x) * (1 - f(x))
  fx = sigmoid(x)
  return fx * (1 - fx)

def mse_loss(y_true, y_pred):
  # y_true and y_pred are numpy arrays of the same length.
  return ((y_true - y_pred) ** 2).mean()

class OurNeuralNetwork:
  '''
  A neural network with:
    - 2 inputs
    - a hidden layer with 2 neurons (h1, h2)
    - an output layer with 1 neuron (o1)

  *** DISCLAIMER ***:
  The code below is intended to be simple and educational, NOT optimal.
  Real neural net code looks nothing like this. DO NOT use this code.
  Instead, read/run it to understand how this specific network works.
  '''
  def __init__(self):
    # Weights
    self.w1 = np.random.normal()
    self.w2 = np.random.normal()
    self.w3 = np.random.normal()
    self.w4 = np.random.normal()
    self.w5 = np.random.normal()
    self.w6 = np.random.normal()

    # Biases
    self.b1 = np.random.normal()
    self.b2 = np.random.normal()
    self.b3 = np.random.normal()

  def feedforward(self, x):
    # x is a numpy array with 2 elements.
    h1 = sigmoid(self.w1 * x[0] + self.w2 * x[1] + self.b1)
    h2 = sigmoid(self.w3 * x[0] + self.w4 * x[1] + self.b2)
    o1 = sigmoid(self.w5 * h1 + self.w6 * h2 + self.b3)
    return o1

  def train(self, data, all_y_trues):
    '''
    - data is a (n x 2) numpy array, n = # of samples in the dataset.
    - all_y_trues is a numpy array with n elements.
      Elements in all_y_trues correspond to those in data.
    '''
    learn_rate = 0.1
    epochs = 1000 # number of times to loop through the entire dataset

    for epoch in range(epochs):
      for x, y_true in zip(data, all_y_trues):
        # --- Do a feedforward (we'll need these values later)
        sum_h1 = self.w1 * x[0] + self.w2 * x[1] + self.b1
        h1 = sigmoid(sum_h1)

        sum_h2 = self.w3 * x[0] + self.w4 * x[1] + self.b2
        h2 = sigmoid(sum_h2)

        sum_o1 = self.w5 * h1 + self.w6 * h2 + self.b3
        o1 = sigmoid(sum_o1)
        y_pred = o1

        # --- Calculate partial derivatives.
        # --- Naming: d_L_d_w1 represents "partial L / partial w1"
        d_L_d_ypred = -2 * (y_true - y_pred)

        # Neuron o1
        d_ypred_d_w5 = h1 * deriv_sigmoid(sum_o1)
        d_ypred_d_w6 = h2 * deriv_sigmoid(sum_o1)
        d_ypred_d_b3 = deriv_sigmoid(sum_o1)

        d_ypred_d_h1 = self.w5 * deriv_sigmoid(sum_o1)
        d_ypred_d_h2 = self.w6 * deriv_sigmoid(sum_o1)

        # Neuron h1
        d_h1_d_w1 = x[0] * deriv_sigmoid(sum_h1)
        d_h1_d_w2 = x[1] * deriv_sigmoid(sum_h1)
        d_h1_d_b1 = deriv_sigmoid(sum_h1)

        # Neuron h2
        d_h2_d_w3 = x[0] * deriv_sigmoid(sum_h2)
        d_h2_d_w4 = x[1] * deriv_sigmoid(sum_h2)
        d_h2_d_b2 = deriv_sigmoid(sum_h2)

        # --- Update weights and biases
        # Neuron h1
        self.w1 -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w1
        self.w2 -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w2
        self.b1 -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_b1

        # Neuron h2
        self.w3 -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w3
        self.w4 -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w4
        self.b2 -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_b2

        # Neuron o1
        self.w5 -= learn_rate * d_L_d_ypred * d_ypred_d_w5
        self.w6 -= learn_rate * d_L_d_ypred * d_ypred_d_w6
        self.b3 -= learn_rate * d_L_d_ypred * d_ypred_d_b3

      # --- Calculate total loss at the end of each epoch
      if epoch % 10 == 0:
        y_preds = np.apply_along_axis(self.feedforward, 1, data)
        loss = mse_loss(all_y_trues, y_preds)
        print("Epoch %d loss: %.3f" % (epoch, loss))
        rafi.append(epoch)
        shafi.append(loss)

# Define dataset
data = np.array([
  [-2, -1],  # Alice
  [25, 6],   # Bob
  [17, 4],   # Charlie
  [-15, -6], # Diana
])
all_y_trues = np.array([
  1, # Alice
  0, # Bob
  0, # Charlie
  1, # Diana
])

# Train our neural network!
network = OurNeuralNetwork()
network.train(data, all_y_trues)
```
